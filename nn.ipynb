```python
#Printing a specific batch 
import itertools
def show_images(images): 
    figure=plt.figure()
    num_of_image=64
    for index in range(1,num_of_image):
        plt.subplot(8,8,index)
        plt.imshow(images[index].numpy().squeeze(), cmap='gray_r')

k=5
my_sample = next(itertools.islice(train_set, k, None))
show_images(my_sample[0])
```
